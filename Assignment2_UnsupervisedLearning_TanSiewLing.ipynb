**# CHOOSING THE NUMBER OF CLUSTERS K**

A higher Silhouette Coefficient score relates to a model with better defined clusters.

Assuming that we don’t know the number of iris clusters initially, we can
use this coefficient (s) to determine the K to use.

```python
# Clustering silhouette_score
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

X = load_iris().data

for n_cluster in range(2, 11):
 kmeans = KMeans(n_clusters=n_cluster).fit(X)
 label = kmeans.labels_
 
 sil_coeff = silhouette_score(X, label, metric='euclidean')
 print("For n_clusters={}, The Silhouette Coefficient is {}".format(n_cluster, sil_coeff))
```

# (B)Clustering Learning Algorithm - DBSCAN 


DBSmodel = DBSCAN(eps=0.9, min_samples=5) 


```python
#DBSCAN

import numpy as np

from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler


# load the iris datasets
dataset = load_iris()
X = dataset.data

#standardize data
X = StandardScaler().fit_transform(X)

# fit a model to the data
DBSmodel=DBSCAN()
#default eps=0.5, leaf_size=30, metric='euclidean', metric_params=None, min_samples=5,
#DBSmodel = DBSCAN(eps=0.5, min_samples=5) #2 cluster
#DBSmodel = DBSCAN(eps=0.5, min_samples=10) #3 cluster
DBSmodel = DBSCAN(eps=0.9, min_samples=5) #3 cluster
DBSmodel.fit(X)
print(DBSmodel)

# Visualize the results
import matplotlib.pyplot as plt
X = dataset.data
y_dbs = DBSmodel.fit_predict(X) #cluster from model

plt.figure()
# scatter plot the petal length (column 2), petal width (column 3)
plt.scatter(X[:, 2], X[:, 3], c=y_dbs, s=50, cmap='viridis')


core_samples_mask = np.zeros_like(DBSmodel.labels_, dtype=bool)
core_samples_mask[DBSmodel.core_sample_indices_] = True
labels = DBSmodel.labels_
print(labels)

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)

#print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
#print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
#print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
#print("Adjusted Rand Index: %0.3f" % metrics.adjusted_rand_score(labels_true, labels))
#print("Adjusted Mutual Information: %0.3f"  % metrics.adjusted_mutual_info_score(labels_true, labels))
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(X, labels))

# #############################################################################
# Plot result
import matplotlib.pyplot as plt

plt.figure()
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

```python
#Gaussian Mixture

from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import numpy as np

#load data
from sklearn import datasets
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# load the iris datasets
dataset = load_iris()
X = dataset.data

#standardize data
X = StandardScaler().fit_transform(X)

#Gaussian Mixture Model
gmm = GaussianMixture(n_components=7)
                
gmm.fit(X)
print(gmm)

# Visualize the results
import matplotlib.pyplot as plt
X = dataset.data
y_gmm = gmm.predict(X) #cluster from model

# scatter plot the petal length (column 2), petal width (column 3)
plt.scatter(X[:, 2], X[:, 3], c=y_gmm, s=50, cmap='viridis')

# show centroid centres as grey circle opacity 50%
#centers = gmm.cluster_centers_ #need not be a data point
#plt.scatter(centers[:, 2], centers[:, 3], c='black', s=200, alpha=0.5)

```

```python
#Gaussian Mixture

from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

#load data
from sklearn import datasets
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

%matplotlib inline



# load the iris datasets
dataset = load_iris()
X = dataset.data

#standardize data
X = StandardScaler().fit_transform(X)

#Gaussian Mixture Model
gmm = GaussianMixture(n_components=3)
gmm.fit(X)

proba_lists = gmm.predict_proba(X)

#Plotting
colored_arrays = np.matrix(proba_lists)
colored_tuples = [tuple(i.tolist()[0]) for i in colored_arrays]
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=colored_tuples, edgecolor="k", s=50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("Gaussian Mixture Model", fontsize=14)
```
